#### 8.Frage
#### Sind bestimmte Schauspieler besonders häufig vertreten?
#### Are certain actors particularly frequently represented?

```python
#Aus den einzelnen Spalten neue Serien machen.
#Make new series from the individual columns.

Cast1=df["Cast1"].copy()
Cast2=df["Cast2"].copy()
Cast3=df["Cast3"].copy()
Cast4=df["Cast4"].copy()
```

```python
#Die neune Serien zusammenfügen.
#Matching the nine series.

casting=pd.concat([Cast1,Cast2,Cast3,Cast4],axis = 0, ignore_index=True)
```

```python
casting
```

```python
#Erstellung eines DataFrames
#Creation of a DataFrame

casting=pd.DataFrame(casting)
```

```python
#Umbenennung des Spaltennamens
#Rename the column name

casting.rename(columns={0:"Schauspieler"},inplace=True)
```

```python
S=casting.pivot_table(index="Schauspieler",aggfunc={"Schauspieler": "count"})
```

```python
#Schauspieler der die meisten Rollen hatte.
#Actor who had the most roles.

S.idxmax()
```

```python
#auf Rangliste
#on ranking list

S.max()
```

```python
S.head(2)
```

```python
S.rename(columns={"Schauspieler":"Anzahl"},inplace=True)
```

```python
# Ersichtlich machen der Rangliste.
# Making the ranking list visible.

S[S["Anzahl"]>=3].plot.bar(figsize=(14, 6),grid=True,rot=45,
                                      xlabel="Name der Schauspieler",ylabel="Anzahl der Filme",
                                     title="Schauspieler die in mindestens drei Filmen mitspielten")
```
