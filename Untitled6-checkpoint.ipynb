# Eingabefenster für die Imobilienbewertung

```python
#Erstellung des Fensters
window = tk.Tk()
#Größe des Fensters
window.geometry("600x600")
#Farbe des Fensters
bg="#c6e2ff"
window.configure(bg="#c6e2ff")
#Name des Fensters
window.title("Imobilienbewertung")
# Kopf- und Fusslabel
kopf_lbl=tk.Label(window,text="Ihre Imobilienbewertung",font=("Helvetica",25,["bold","underline"]),fg="#005aff",bg="#c6e2ff")
# Eingabe Label
grundstueck_lbl=tk.Label(window, text="Wie gross ist das Grundstück in m²?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
wohnflaeche_lbl=tk.Label(window, text="Wie gross ist die Wohnfläche in m²?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
architekt_lbl=tk.Label(window, text="Handelt es sich um ein Architektenhaus?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
makler_lbl=tk.Label(window, text="Soll das Haus von einem Makler verkauft werden?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
denkmalsch_lbl=tk.Label(window, text="Steht das Haus ganz oder teilweise unter Denkmalschutz?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
baujar_lbl=tk.Label(window, text="In welchem Jahr wurde das Haus gebaut?", font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
wo_lbl=tk.Label(window, text="Wo befindet sich das Haus?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
ausstattung_lbl=tk.Label(window, text="Zustand der Immobilie.",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
hausart_lbl=tk.Label(window, text="Welche Hausart hat die Immobilie?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
bundesland_lbl=tk.Label(window, text="in welchem Bundesland steht das Haus?",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
ausgabe_lbl=tk.Label(window,text="Der zu erwartende Verkaufpreis:",font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
```

```python
# Textfelder
grundstueck_txt=tk.Entry(window,font=("Helvetica",12)) #für Grundstückfläche
wohnflaeche_txt=tk.Entry(window,font=("Helvetica",12)) #für Wohnläche
baujahr_txt=tk.Entry(window,font=("Helvetica",12)) # Eingabe des Baujahres
```

```python
# Listen zu Eingaben
bundesland_lst=["Baden-Württemberg","Bayern","Berlin","Brandenburg","Bremen","Hamburg","Hessen","Mecklenburg-Vorpommern",
            "Niedersachsen","Nordrhein-Westfalen","Rheinland-Pfalz","Saarland","Sachsen","Sachsen-Anhalt",
            "Schleswig-Holstein","Thüringen"]
wo_lst=["Stadt","Land"]
ausstattung_lst=["Rohbau","Sanierungsbedarf","Renovierungsbedarf","Einfach","Gehoben"]
hausart_lst=["Einfamilienhaus","Doppelhaushälfte","Mehrfamilienhaus"]
architekt_lst=["Ja","Nein"]
makler_lst=["Ja","Nein"]
denkmalsch_lst=["Ja","Nein"]

# Eingaben
architekt_cobox = ttk.Combobox(window,values=architekt_lst)
makler_cobox = ttk.Combobox(window,values=makler_lst)
denkmalsch_cobox = ttk.Combobox(window,values=denkmalsch_lst)
wo_cobox = ttk.Combobox(window,values=wo_lst)
ausstattung_cobox = ttk.Combobox(window,values=ausstattung_lst)
hausart_cobox = ttk.Combobox(window,values=hausart_lst)
bundesland_cobox = ttk.Combobox(window,values=bundesland_lst)

# Ausgabe
ausgabe = tk.Label(window,font=("Helvetica",12,'bold'),fg="#005aff",bg="#c6e2ff")
```

```python
def rechner():
    try:       
        grundstueck_n=int(grundstueck_txt.get())
        wohnflaeche_n=int(wohnflaeche_txt.get())
        baujahr_n=int(baujahr_txt.get())
        
        architekt=architekt_cobox.get()
        makler=makler_cobox.get()
        denkmalsch=denkmalsch_cobox.get()
        wo=wo_cobox.get()
        ausstattung=ausstattung_cobox.get()
        hausart=hausart_cobox.get()
        bundesland=bundesland_cobox.get()
        
        
        
        year_now=int(date.today().strftime("%Y"))
        baujahr_ag=1 if baujahr_n == year_now else (year_now-baujahr_n)*(0.1/100)
    
        architekt_ag=1+df_Kostenfaktoren.loc["Geplant von Architekt"][0]/100 if architekt =="Ja" else 1
        makler_ag=1+df_Kostenfaktoren.loc["Makler"][0]/100 if makler =="Ja" else 1
        denkmalsch_ag=1-df_Kostenfaktoren.loc["Denkmalschutz"][0]/100 if denkmalsch =="Ja" else 1
        
        
        
        if wo == "Stadt":
            wo_ag=df_StadtLand.loc["Stadt"][0]
        else:
            wo_ag=df_StadtLand.loc["Land"][0]
    
        ausstattung_ag=df_Ausstattung.loc[ausstattung][0]
        hausart_ag=df_Hausart.loc[hausart][0]
        bundesland_ag=df_Bundesland.loc[bundesland][0]
        grundstueck_ag=df_Kostenfaktoren.loc["QM-Grundstück"][0]
        wohnflaeche_ag=df_Kostenfaktoren.loc["QM-Wohnfläche"][0]
    
    except:
        pass
    
        
    
    berechnung=(grundstueck_n*grundstueck_ag + wohnflaeche_n*wohnflaeche_ag)*architekt_ag*makler_ag*denkmalsch_ag*baujahr_ag
                
    message = str(round(berechnung,3))+" €"
    ausgabe.configure(text=message)  
```

```python
button1=tk.Button(window,text="Berechnung",command=rechner)
button2=tk.Button(window,text="Beenden",command=window.destroy)
```

```python
# Kopf
kopf_lbl.pack()
# Grundstück
grundstueck_lbl.pack()#.place(x=20,y=50)
grundstueck_txt.pack()#.place(x=320,y=50)
# Wohnfläche
wohnflaeche_lbl.pack()#.place(x=20,y=80)
wohnflaeche_txt.pack()#.place(x=320,y=80)
# Baujahr
baujar_lbl.pack()
baujahr_txt.pack()
# Architektenhaus
architekt_lbl.pack()
architekt_cobox.pack()
# Makler
makler_lbl.pack()
makler_cobox.pack()
# Denkmalschutz
denkmalsch_lbl.pack()
denkmalsch_cobox.pack()
# Land oder Stadt
wo_lbl.pack()
wo_cobox.pack()
# Ausstattung
ausstattung_lbl.pack()
ausstattung_cobox.pack()
# Hausart
hausart_lbl.pack()
hausart_cobox.pack()
# Bundesland
bundesland_lbl.pack()
bundesland_cobox.pack()
# Button
button1.pack()
button2.pack(side=RIGHT, padx=15, pady=15)
# Ausgbe
ausgabe_lbl.pack()
ausgabe.pack()
```
